```python
import matplotlib.pyplot as plt
loss = [12.97, 3.98, 2.58, 2.15, 1.54, 1.16, 1.11, 1.08, 0.77, 0.82]
accuracy = [96.02, 98.80, 99.20, 99.31, 99.48, 99.65, 99.65, 99.65, 99.73, 99.72]
epochs = 10

# Create plots
plt.figure(figsize=(12, 5))

# Plot Loss
plt.subplot(1, 2, 1)
plt.plot(range(1, epochs + 1), loss, label='Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot Accuracy
plt.subplot(1, 2, 2)
plt.plot(range(1, epochs + 1), accuracy, label='Accuracy')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.suptitle('CNN using Keras', fontsize=16)

plt.tight_layout()
plt.show()
```
